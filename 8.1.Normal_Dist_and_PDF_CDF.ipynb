```python
def get_pdf_probability(dataset,startRange,endRange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax=sns.distplot(dataset,kde=True,kde_kws={'color':'Yellow'},color='Green')
    pyplot.axvline(startRange,color='Red')
    pyplot.axvline(endRange,color='Red')
    #Generate sample
    sample=dataset
    #calculate parameters
    mean=sample.mean()
    std=sample.std()
    print("Mean=%.3f, Standard Deviation=%.3f" %(mean,std))
    #Define distribution
    dist=norm(mean,std)
    #sample probabilities for a range of outcomes
    values=[value for value in range(startRange,endRange)]
    prob=[dist.pdf(value) for value in values]
    probabilities=sum(prob)
    print("The area between range({},{}):{}".format(startRange,endRange,probabilities))
    return probabilities
```

```python
get_pdf_probability(dataset['hsc_p'],60,70)
```
